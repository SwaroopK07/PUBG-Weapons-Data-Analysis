### Visualize the Data

```python
plt.figure(figsize = (10,10))
data=df.copy()
data.groupby('Weapon Type')['Damage'].mean().plot.bar()
plt.title("Weapon Type")
plt.xlabel("Weapon Type")
plt.ylabel("Damage")
plt.show()
```

### From above graph we can say that shotguns has more avereage damage and pistol has less average damage

```python
plt.figure(figsize=(10,10))
data = df.copy()
data.groupby('Weapon Type')['Magazine Capacity'].mean().plot.bar()
plt.xlabel('Weapon Type')
plt.ylabel('Magazine Capacity')
plt.title('Weapon Type')
plt.show()
```

###  Average magzine capacity of light machine gun is more and average magazine capacity of shotgun is less

```python
data = df.copy()
plt.figure(figsize=(10,10))
data.groupby('Weapon Type')['Range'].mean().plot.bar()
plt.xlabel("Weapon Type")
plt.ylabel("Range (m)")
plt.title("Weapon Type")
plt.show()
```

### Average range of bullet travel is more for sniper rifle and less for shotgun and pistol

```python
plt.figure(figsize=(10,10))
data = df.copy()
data.groupby('Weapon Type')['Rate of Fire'].mean().plot.bar()
plt.xlabel("Weapon Type")
plt.ylabel("Rate of Fire")
plt.title("Weapon Type")
plt.show()
```

### Rate of fire is more for sniper rifle and less for Designed Marksman Rifle

```python
plt.figure(figsize=(10,10))
data=df.copy()
data.groupby("Weapon Type")["Shots to Kill (Chest)"].mean().plot.bar()
plt.xlabel("Weapon Type")
plt.ylabel("Shots to Kill (Chest)")
plt.title("Weapon Type")
plt.show()
```

### If you are firing on chest then average number bullets which takes enemy down is greater for pistol and less for shotgun(but in close range).

```python
plt.figure(figsize=(10,10))
data=df.copy()
data.groupby('Weapon Type')['Damage Per Second'].mean().plot.bar()
plt.xlabel('Weapon Type')
plt.ylabel('Damage Per Second')
plt.title('Weapon Type')
plt.show()
```

### Average Damage per second is more for shotgun and less for sniper rifle
